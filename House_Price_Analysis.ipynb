## capping and flooring for outlier treatment

```python
np.percentile(house_df.n_hot_rooms,[99])
```

```python
uv=np.percentile(house_df.n_hot_rooms,[99])[0]
```

```python
uv
```

```python
house_df[house_df.n_hot_rooms>uv]
```

```python
house_df.n_hot_rooms[house_df.n_hot_rooms>3*uv]=3*uv
```

```python
lv=np.percentile(house_df.rainfall,[1])[0]
```

```python
lv
```

```python
house_df.rainfall[(house_df.rainfall<0.3*lv)]=0.3*lv
```

```python
house_df[(house_df.rainfall<lv)]
```

##### Ridge regression and lasso shrinkage techniques
first need to standardizing the data

```python
from sklearn import preprocessing 
```

```python
scaler=preprocessing.StandardScaler().fit(x_train)
```

```python
x_train_s=scaler.transform(x_train)
```

```python
x_test_s=scaler.transform(x_test)
```

```python
from sklearn.linear_model import Ridge
```

```python
lr_ridge=Ridge(alpha=0.6)
lr_ridge.fit(x_train_s,y_train)
```

```python
r2_score(y_test,lr_ridge.predict(x_test_s))
```

```python
from sklearn.model_selection import validation_curve
```

```python
param_range=np.logspace(-2,8,100)
param_range
```

```python
train_scores,test_scores=validation_curve(Ridge(),x_train_s,y_train,'alpha',param_range,scoring='r2')
```

```python
train_mean=np.mean(train_scores,axis=1)
```

```python
test_mean=np.mean(test_scores,axis=1)
```

```python
max(test_mean)
```

```python
sns.jointplot(x=np.log(param_range),y=test_mean) #plot between lambda and test_means
```

```python
np.where(test_mean==max(test_mean))
```

```python
param_range[30]
```

```python
lr_ridge_best=Ridge(alpha=param_range[30])
```

```python
lr_ridge_best.fit(x_train_s,y_train)
```

```python
r2_score(y_train,lr_ridge_best.predict(x_train_s))
```

```python
r2_score(y_test,lr_ridge_best.predict(x_test_s))
```
